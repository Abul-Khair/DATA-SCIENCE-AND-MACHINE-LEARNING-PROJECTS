```python
x = df.drop(columns='current price')
y = df['current price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)

lr = LinearRegression()
lr.fit(x_train_scaled,y_train)


y_pred = lr.predict(x_test_scaled)

from sklearn.metrics import r2_score, root_mean_squared_error, mean_squared_error

r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)


print(f"Accuracy{r2} \n MSE:{mse} \n RMSE:{rmse}")
diff = pd.DataFrame(np.c_[y_test, y_pred],columns= ["Actual price","Predicted Price"])
diff
```

```python
plt.figure(figsize=(10,5))
sns.distplot(y_test,label="Actual price")
sns.distplot(y_pred,label="Predicted price")
plt.title("Actual vs predicted price")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(x_train_scaled,y_train)


y_pred = rfr.predict(x_test_scaled)

from sklearn.metrics import r2_score, root_mean_squared_error, mean_squared_error

r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)


print(f"Accuracy{r2} \n MSE:{mse} \n RMSE:{rmse}")
diff = pd.DataFrame(np.c_[y_test, y_pred],columns= ["Actual price","Predicted Price"])
diff
```

```python
plt.figure(figsize=(10,5))
sns.distplot(y_test,label="Actual price")
sns.distplot(y_pred,label="Predicted price")
plt.title("Actual vs predicted price")
plt.legend()
plt.show()
```
